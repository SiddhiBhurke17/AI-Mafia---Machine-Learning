# MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
```

## __Basic Plotting__

#### Scatter vs Plot points

```python
###specify data to plot
x = [18 , 19 , 21, 23]
y = [3 , 7 , 9 ,1]
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

#### Setting x and y axis range

```python
plt.ylim(1,8)###defines max of x axis
plt.xlim(1,8)### defines max of x axis
###range only defines the max of axes for plotting u need to give co ornt of x y but in given range for visible graph"
x = [1 , 2]
y= [7 , 8 ]
plt.plot(x,y)
plt.show()
plt.close()
```

## __Labeling the graph__

```python
####plt.xlabel('X-axis')___ ## or whatever you want eg age of students
plt.xlabel('Age of students')
####plt.ylabel('Y-axis')___ ## or whatever you want eg NO of students
plt.ylabel('No. of students')
plt.plot(x,y)
plt.title('Coding Club')
plt.show()
```

## __Plot 2 lines on same plot  and Styling the plots__

#### Plot 2 lines with given style

```python
##line 1 points
x1 = [1,2,3]
y1 = [2,4,5]
#line 2 points
x2 = [2,6,7]
y2 = [8,5,3]
plt.plot(x1,y1,label = 'line1')
#plt.style.use()
plt.plot(x2,y2,label = 'line2')
plt.style.use( 'seaborn-v0_8-dark-palette')
plt.legend()
plt.show()
```

#### Styles available

```python
plt.style.available   ######style is used above
```

#### another way to style the plot

```python
x = [18 , 19 , 21, 23]
y = [3 , 7 , 9 ,1]
plt.plot(x,y,color='green',linestyle='dashed', linewidth =3 , marker = '^',markerfacecolor='blue',markersize = 8)
```

# Plotting Bar chart

```python
# x cordnt of left sides of bars
left = [1,2,5, 7,15,16]
#heigth of bars
height = [10 , 24 ,20 , 36 ,40 ,6]
#labels for bars
tick_label = ['one' , 'two' , 'three', 'four','five' ,'six']
##plotiing bar chart
plt.bar(left , height , tick_label=tick_label,width=0.8,color=['red','blue'])
plt.show()
```

# Plotting histogram

```python
#import matplotlib.pyplot as plt

# Data
age = [10, 11, 12, 13]
people = [333, 7, 9, 1]

# Histogram settings
range = (0, 100)
bins = 7

# Plot histogram
plt.hist(age, bins=bins, range=range, color='red', histtype='bar', rwidth=0.05)  # Adjusted rwidth value
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Ages')
plt.show()
```

# Plot a Scatter Plot

```python
x = [1,2,3,4,5]
y = [2,3,4,5,7]
plt.scatter(x,y,label="stars",color="blue",marker="*",s=30)
```

# Plot a Piechart

```python
#define labels
activities = ['eat', 'sleep','work','play']
#portion covered by each label
slices = [3,7,5,9]
#colour for each slices
colors=['r','m','g','b']
###___plotting the piechart____###
plt.pie(slices,labels=activities,colors=colors,startangle=90,shadow=True,explode=(0,0,0.05,0),radius=1.2,autopct='@1.1f%%')
```

# Plotting curves of given equation

```python
x = np.arange(0,2*(np.pi),0.1)
y = np.sin(x)
plt.plot(x,y)
```
